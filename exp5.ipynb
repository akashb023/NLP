```python
import nltk

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
para = """Bluetooth is a short-range wireless technology standard that is used for exchanging data between fixed and mobile devices over short distances and building personal area networks (PANs). ... enhance IoT capabilities."""
```

```python
sentence = nltk.sent_tokenize(para)
sentence
```

```python
stemmer = PorterStemmer()

for i in range(len(sentence)):
    words = nltk.word_tokenize(sentence[i])
    words = [stemmer.stem(word) for word in words]
    sentence[i] = ' '.join(words)

sentence
```

```python
para1 = para
sentence1 = nltk.sent_tokenize(para1)

lemmatizer = WordNetLemmatizer()

for i in range(len(sentence1)):
    words = nltk.word_tokenize(sentence1[i])
    newwords = [lemmatizer.lemmatize(word) for word in words]
    sentence1[i] = ' '.join(newwords)

sentence1
```

```python
para2 = para
sentences2 = nltk.sent_tokenize(para2)

for i in range(len(sentences2)):
    words = nltk.word_tokenize(sentences2[i])
    newwords = [word for word in words if word.lower() not in stopwords.words('english')]
    sentences2[i] = ' '.join(newwords)

sentences2
```

```python
para3 = para
word = nltk.word_tokenize(para3)

tag_word = nltk.pos_tag(word)
tag_word
```

```python
named_Ent = nltk.ne_chunk(tag_word)
print(named_Ent)
```
